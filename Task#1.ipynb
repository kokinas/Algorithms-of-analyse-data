1. –ü–æ–¥–±–µ—Ä–∏—Ç–µ —Å–∫–æ—Ä–æ—Å—Ç—å –æ–±—É—á–µ–Ω–∏—è (eta) –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π

```python
import numpy as np
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
n = X.shape[0]
eta = 0.01
n_iter = 1000
W = np.array([2., 0.1])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')
for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = np.mean((y - y_pred)**2)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        eta /= 1.01
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

2*. –í —ç—Ç–æ–º –∫–æ–¥–µ –º—ã –∏–∑–±–∞–≤–ª—è–µ–º—Å—è –æ—Ç –∏—Ç–µ—Ä–∞—Ü–∏–π –ø–æ –≤–µ—Å–∞–º, –Ω–æ —Ç—É—Ç –µ—Å—Ç—å –æ—à–∏–±–∫–∞, –∏—Å–ø—Ä–∞–≤—å—Ç–µ –µ–µ

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
n = X.shape[0]
eta = 0.01 
n_iter = 500
W = np.array([4., 10.])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    W -= eta * (1/n * 2 * np.dot(X.T, (y_pred - y)))    # X.T –≤–º–µ—Å—Ç–æ X
    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,2)}')
```

3*. –í–º–µ—Å—Ç–æ —Ç–æ–≥–æ, —á—Ç–æ–±—ã –∑–∞–¥–∞–≤–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π, –∑–∞–¥–∞–π—Ç–µ –¥—Ä—É–≥–æ–µ —É—Å–ª–æ–≤–∏–µ –æ—Å—Ç–∞–Ω–æ–≤–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ - –∫–æ–≥–¥–∞ –≤–µ—Å–∞ –ø–µ—Ä–µ—Å—Ç–∞—é—Ç –∏–∑–º–µ–Ω—è—Ç—å—Å—è –º–µ–Ω—å—à–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ –ø–æ—Ä–æ–≥–∞  ùúñ
 .

```python
n = X.shape[0]
eta = 0.01
e = 0.01
W = np.array([10., 9.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')
delta_err = 1000
err = 100000
i = 0
while delta_err > e:
    y_pred = np.dot(X, W)
    last_err = err
    err = calc_mse(y, y_pred)
    delta_err = last_err - err 
    W -= eta * (1/n * 2 * np.dot(X.T, (y_pred - y)))    # X.T –≤–º–µ—Å—Ç–æ X
    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,2)}, delta_err={delta_err}')
    i+=1
```
